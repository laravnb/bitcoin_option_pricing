```python
import numpy as np
import matplotlib.pyplot as plt   #other math package??
                                  
def monte_carlo_paths(S, T, r, sigma, n, N):   #n steps,  N?

    #Output 
    #[steps,N] Matrix of asset paths
    dt= T/n
    MC = np.log(S) + np.cumsum(((r - sigma**2/2)*dt + sigma*np.sqrt(dt)* 
                                np.random.normal(size=(n,N))),axis=0)
    
    return np.exp(MC)

#Kelly
#i=0
S = 100 #S_input[i] #stock price S_{0}
K = 110 #K_input # strike
T = 1/2 #T_input[i] # time to maturity
r = 0.05 #r_input # risk free risk in annual %
q = 0
sigma = 0.25 #sigma_input[i] # annual volatility in %
n = 100 # time steps
N = 10000 # number of trials

monte_carlo= monte_carlo_paths(S,T,r,sigma,n,N)

plt.plot(monte_carlo);
plt.xlabel("Time Increments")
plt.ylabel("Stock Price")
plt.title("Geometric Brownian Motion")
    
#My code
#repl = 100
#G = 0
#C = 0
#summ = 0

#for i in repl:
#    G = max(0, Si - Ki)
#    C = exp(-r * Ti)  * G
#    summ = summ + C
    
#CM = summ / repl
```
